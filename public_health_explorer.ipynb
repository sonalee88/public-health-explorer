```python
# ⬛️ Cell 1: Setup
!pip install pandas matplotlib seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for plots
sns.set(style="whitegrid")
```

```python
# ⬛️ Cell 2: Upload CSV File
from google.colab import files
uploaded = files.upload()
```

```python
# ⬛️ Cell 3: Load Dataset
import io

# Replace with your actual file name if different
df = pd.read_csv(io.BytesIO(uploaded['NPHA-doctor-visits.csv']))

# Preview the dataset
df.head()
```

```python
# ⬛️ Cell 4: Dataset Info
df.info()
df.describe()
```

```python
# ⬛️ Cell 5: Average Age of Patients
avg_age = df['Age'].mean()
print(f"Average Age of Patients: {avg_age:.2f}")
```

```python
# ⬛️ Cell 6: Distribution of Patients by Gender
gender_dist = df['Gender'].value_counts()
print(gender_dist)
gender_dist.plot(kind='bar', title='Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# ⬛️ Cell 7: Average Health Score (Physical + Mental + Dental / 3)
df['Average Health Score'] = df[['Phyiscal Health', 'Mental Health', 'Dental Health']].mean(axis=1)
avg_score = df['Average Health Score'].mean()
print(f"Average Health Score: {avg_score:.2f}")
```

```python
# ⬛️ Cell 8: Patients Reporting Trouble Sleeping
# 1 means trouble sleeping, 0 means no
trouble_sleep = df[df['Trouble Sleeping'] == 1].shape[0]
print(f"Number of Patients Reporting Trouble Sleeping: {trouble_sleep}")
```

```python
# ⬛️ Cell 9: Relationship Between Age and Health Score
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Age', y='Average Health Score')
plt.title('Age vs Average Health Score')
plt.xlabel('Age')
plt.ylabel('Health Score')
plt.grid(True)
plt.show()
```

```python
# ⬛️ Cell 10: Patients by Race
race_dist = df['Race'].value_counts()
print("Patients by Race:\n", race_dist)

race_dist.plot(kind='bar', title='Patients by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# ⬛️ Cell 11: Average Number of Doctors Visited
avg_visits = df['Number of Doctors Visited'].mean()
print(f"Average Number of Doctors Visited: {avg_visits:.2f}")
```

```python
# ⬛️ Cell 12: Save Summary Report
avg_age = df['Age'].mean()

with open("summary_report.txt", "w") as f:
    f.write(f"Average Age: {avg_age:.2f}\n")
    f.write(f"Average Health Score: {avg_score:.2f}\n")
    f.write(f"Patients w/ Trouble Sleeping: {trouble_sleep}\n")
    f.write(f"Average No. of Doctors Visited: {avg_visits:.2f}\n")

print("✅ Summary report saved to `summary_report.txt`.")
```
